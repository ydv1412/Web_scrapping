```python
import numpy as np
from sklearn import datasets,neighbors
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
```

```python
def knn_on_datasets(data , n_neighbors = 15):
    
    x = data[ : , :2]
    y = data[ : , 2]
    
    h = 0.02
    color_light = ListedColormap(['#FFFFAA','#AAFFFF'])
    color_bold  = ListedColormap(['#FF0000','#0000FF'])
    
    clf = neighbors.KNeighborsClassifier(n_neighbors)
    clf.fit(x,y)
    
    x_min,x_max = x[:,0].min()-1 , x[:,0].max()+1 
    y_min,y_max = x[:,1].min()-1 , x[:,1].max()+1
    
    xx,yy = np.meshgrid(np.arange(x_min , x_max ,h),np.arange(y_min , y_max ,h))
    
    z = clf.predict(np.c_[xx.ravel(),yy.ravel()])
    
    z = z.reshape(xx.shape)
    
    plt.figure()
    plt.pcolormesh( xx, yy, z, cmap = color_light)
    
    plt.scatter( x[:,0] , x[:,1] , c=y , cmap = color_bold)
  
    
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    
    plt.title('k_value ='+str(n_neighbors))
    
    plt.show()
```

```python
data = np.genfromtxt('demo_data/6.overlap.csv',delimiter = ',')
knn_on_datasets(data,1)
knn_on_datasets(data,5)
knn_on_datasets(data,15)
knn_on_datasets(data,30)
knn_on_datasets(data,50)
```

```python
knn_on_datasets(data,100)
```
